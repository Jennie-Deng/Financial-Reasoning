```python
import pandas as pd
```

```python
models=[
        "gpt_o1",
        "Gemini_Pro1",
        "deepseek",
        "llama",
        "gpt_4o",
        "gemini_1.5_pro",
        "qwen",
        "Llava",
        "Llama_3.2_vision",
        "qwen2_vl",
        ]
def file_path (model):
    filePath=f"/Users/sden118/Desktop/FinReasoning/errorLog/{model}_ErrorLog.json"
    return filePath
```

```python
errortpype=pd.read_csv("/Users/sden118/Desktop/FinReasoning/evaluation/errorType.csv")
errortpype=errortpype[["ID","Error"]]
errortpype
```

```python
df=pd.DataFrame()
for model in models:
    filePath=file_path(model)
    df1=pd.read_json(filePath)
    df1=df1[df1['Model Answer']=="Answer not found"]
    df1['Model']=model
    df=pd.concat([df,df1], ignore_index=True)
df
#df.to_excel("//Users/sden118/Desktop/FinReasoning/evaluation/AnswerNotFound.xlsx",index=False)
```

```python
merged_df=pd.merge(errortpype,df,on="ID",how="left")
merged_df
```

```python
from PIL import Image
merged_df["Image"][1]
```
